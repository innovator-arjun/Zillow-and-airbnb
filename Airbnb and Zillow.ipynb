```python
airbnbdf['price'] = airbnbdf['price'].str.replace('$', '')
airbnbdf['price'] = airbnbdf['price'].str.replace(',', '')
airbnbdf['weekly_price'] = airbnbdf['weekly_price'].str.replace('$', '')
airbnbdf['weekly_price'] = airbnbdf['weekly_price'].str.replace(',', '')
airbnbdf['monthly_price'] = airbnbdf['monthly_price'].str.replace('$', '')
airbnbdf['monthly_price'] = airbnbdf['monthly_price'].str.replace(',', '')
airbnbdf['cleaning_fee'] = airbnbdf['cleaning_fee'].str.replace('$', '')
airbnbdf['cleaning_fee'] = airbnbdf['cleaning_fee'].str.replace(',', '')
```

```python
airbnbdf[['price','weekly_price','monthly_price','cleaning_fee']]=airbnbdf[['price','weekly_price','monthly_price','cleaning_fee']].astype(float)
```

```python
airbnbdf.weekly_price.fillna(airbnbdf.price*7, inplace=True)
```

```python
airbnbdf.monthly_price.fillna(airbnbdf.price*30, inplace=True)
```

```python
airbnbdf['cleaning_fee'].fillna((airbnbdf['cleaning_fee'].mean()), inplace=True)
```

```python
occupancy=0.75
airbnbdf['revenue_daily']=((airbnbdf.price*occupancy)+(airbnbdf.cleaning_fee*occupancy))
```

```python
airbnbdf['revenue_weekly']=((airbnbdf.weekly_price*occupancy)+(airbnbdf.cleaning_fee*occupancy*7))
```

```python
airbnbdf['revenue_monthly']=((airbnbdf.monthly_price*occupancy)+(airbnbdf.cleaning_fee*occupancy*30))
```

```python
airbnbdf['revenue_yearly']=((airbnbdf.monthly_price*occupancy*12)+(airbnbdf.cleaning_fee*occupancy*365))
```
